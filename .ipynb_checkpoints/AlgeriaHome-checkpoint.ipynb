```python
from selenium import webdriver 
from selenium.webdriver.chrome.service import Service as ChromeService 
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import requests
import json
from bs4 import BeautifulSoup as soup
import time
```

```python
base_url = "https://www.algeriahome.com/search"

# Category a vendre, it can be also "a-louer", to insert into the url, add: "category,{category}"
category = 'a-vendre'        

# Filtering using price range, to use it in the url, add:"sPriceMax,{MaxPrice} (same thing for MinPrice)
# MinPrice = '900'
# MaxPrice = '7200'

# Filtering using ascending and descending sorting
# ascending_sort = "sOrder,i_price/iOrderType,asc"
# descending_sort = "sOrder,i_price/iOrderType,desc"

# Filtering using region, to insert it into the url, add: "region,{region}"
# region = 'Constantine'

# If you want to add multiple filterings, separate them only by "/"
url = f"{base_url}/category,{category}"
```

```python
options = webdriver.ChromeOptions() 
options.headless = True

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options) 
driver.get(url)
# Define a function to scroll to the bottom of the page
def scroll_to_bottom(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        # Wait for the page to load
        time.sleep(6)  # Adjust the sleep time as needed
        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height


# Scroll to the bottom of the page to load all content
scroll_to_bottom(driver)

parent_links = []


try:
    while True:
        # Scrape links on the current page
        wait = WebDriverWait(driver, 10)
        parent_elements = wait.until(EC.visibility_of_all_elements_located((By.XPATH, "//h4/a")))
        parent_links.extend([parent_element.get_attribute("href") for parent_element in parent_elements])

        # Check for the next button
        next_button = driver.find_element(By.XPATH,'//a[contains(@class, "searchPaginationNext")]')
        next_button_href = next_button.get_attribute('href')
    
        # Click on the next button
        driver.get(next_button_href)

        # Wait for the page to load
        wait.until(EC.staleness_of(parent_elements[0]))  # Wait for previous elements to become stale
        scroll_to_bottom(driver)

except NoSuchElementException:
    print("Reached the last page !")

# Scrape data from each element
listings_data = [] 

for listing_link in parent_links: 
    driver.get(listing_link)
    time.sleep(3)    
    try:
        driver.execute_script("window.scrollBy(0, 600);")
        time.sleep(1)
        
        listing_description = driver.find_element(By.XPATH, '//*[@id="description"]/p').text        
        listing_price = driver.find_element(By.XPATH, '//*[@id="item-content"]/ul/li[1]').text
        listing_title = driver.find_element(By.XPATH, '//*[@id="item-content"]/h1/strong').text
        listing_location = driver.find_element(By.XPATH, '//*[@id="item_location"]/li').text
        listing_date = driver.find_element(By.XPATH, '//*[@id="item-content"]/ul/li[2]').text
        try:
            thumbs_images = driver.find_elements(By.CLASS_NAME, 'thumbs-image')
            if not thumbs_images:
                raise NoSuchElementException
                  
            # List to store the href attributes
            href_list = []
              
            # Iterate over each thumbs-image div
            for thumbs_image in thumbs_images:
                # Find the <a> tag within the thumbs-image div
                a_tag = thumbs_image.find_element(By.TAG_NAME, 'a')
                # Get the href attribute value and append to the list
                href_list.append(a_tag.get_attribute('href'))
                  
            # Print the list of href attributes
            print(href_list)
          
        except NoSuchElementException:
            print("No images were found!")
       
            
    except:
        print("Error")
    
    temporary_listings_data = { 
        "title": listing_title,
        "location": listing_location,
        "price": listing_price,
        "description": listing_description,
        "images": href_list,
        "Date": listing_date,
        "link": listing_link,
    } 
    listings_data.append(temporary_listings_data) 
    print(temporary_listings_data)

driver.quit()
```

```python
with open('AlgeriaHome.json', 'w', encoding='utf-8') as json_file:
    json.dump(listings_data, json_file, ensure_ascii=False, indent=4)
```
